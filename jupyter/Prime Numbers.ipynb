# Prime Number Exercises

Not quite [code golf](https://en.wikipedia.org/wiki/Code_golf) but close.


```python
import math
import time

def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0 and n > 2:
        return False
    for i in range(3, int(pow(n, 0.5))):
        if n % i == 0:
            return False
    return True

def main():
    primes = []
    start_time = time.time()

    for a in range(0, 100):
        b = is_prime(a)
        if b:
            primes.append(a)
        print a, 'is prime: ', b

    elapsed_time = time.time() - start_time
    print "Just the primes: ", primes
    print "Elapsed time: %f s" % elapsed_time

if __name__ == '__main__':
    main()
```
